### Building Visualization in Pandas

```python
# import pandas Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\user\Documents\explorer tutorial\xlsx\Apocolypse Food Prep - Bins Lists Tutorial.xlsx")
df = df.set_index('Product ID')
df
```

```python
# To check the different styles of plot available
print(plt.style.available)
plt.style.use('dark_background')
```

```python
# Histogram
df.plot(kind = 'hist')
plt.style.use('dark_background')
```

```python
# Line plot is the default plot for data in python
df.plot()
plt.style.use('classic')
```

```python
df.plot(kind = 'line')
plt.style.use('grayscale')
```

```python
# Stacked Bar Chart
df.plot(kind = 'bar', stacked = 'True', title = 'Product Rating', xlabel = 'Product Identity', ylabel = 'Amount')
plt.style.use('Solarize_Light2')
```

```python
# Bar Chart and Stacked Bar Chart
df.plot(kind = 'bar', title = 'Product Rating', xlabel = 'Product Identity', ylabel = 'Amount')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
# To customize Title
df.plot(kind = 'line', title = 'Product Rating', xlabel = 'Product Identity Code', ylabel = 'Amount')
```

```python
# You can plot for just one parameter by specifying
df['Price'].plot(kind = 'bar', title = 'Product Rating', xlabel = 'Product Identity', ylabel = 'Amount')
```

```python
df['Production Cost'].plot(kind = 'bar', title = 'Product Rating', xlabel = 'Product Identity', ylabel = 'Amount')
```

```python
# Horizontal Barchart
df.plot.barh(stacked = 'True', title = 'Product Rating', xlabel = 'Product Identity', ylabel = 'Amount')
```

```python
df.plot.scatter(x = 'Price', y = 'Production Cost', s = 100, c = 'red', title = 'Product Rating')
```

```python
# Histogram by specifying bins
df.plot.hist(bins = 20)
```

```python
# Box Plot
df.boxplot()
```

```python
# Area Plot
df.plot.area()
```

```python
# To increase the size of area plot we use figsize
df.plot.area(figsize = (10, 5))
```

```python
# Pie Chart
df.plot.pie(y = 'Price', figsize = (12, 6))
```

```python
df.plot.pie(y = 'Production Cost')
```
